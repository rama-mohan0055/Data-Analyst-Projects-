```python
Q1 = df['Vol.'].quantile(0.25)
Q3 = df['Vol.'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
```

```python
import matplotlib.pyplot as plt

# Example data
trading_volume = [1000, 2000, 1500, 3000, 2500]
percentage_change = [1.2, -0.5, 0.8, -1.0, 1.5]

# Create scatter plot
plt.scatter(trading_volume, percentage_change, color='blue', alpha=0.7)
plt.title('Scatter Plot: Trading Volume vs Percentage Price Change')
plt.xlabel('Trading Volume')
plt.ylabel('Percentage Price Change')
plt.grid(True)
plt.show()
```
