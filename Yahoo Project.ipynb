```python
data = {
    "Date": ["28-04-2023", "27-04-2023", "26-04-2023"],
    "Open": [33797, 33382, 33596],
    "High": [34105, 33860, 33646],
    "Low": [33728, 33375, 33236],
    "Close": [34098, 33826, 33302],
    "Adj Close": [34098, 33826, 33302],
    "Volume": [354310000, 343240000, 321170000]}
plt.figure(figsize=(8, 6))
plt.boxplot(data['High'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue', color='black'), 
            whiskerprops=dict(color='black'), 
            flierprops=dict(marker='o', color='red', markersize=8))
plt.xlabel('High Price')
plt.title('Box Plot for High Prices')
plt.show()
```

```python
data = {
    "Date": ["28-04-2023", "27-04-2023", "26-04-2023"],
    "Open": [33797, 33382, 33596],
    "High": [34105, 33860, 33646],
    "Low": [33728, 33375, 33236],
    "Close": [34098, 33826, 33302],
    "Adj Close": [34098, 33826, 33302],
    "Volume": [354310000, 343240000, 321170000]}
data = pd.DataFrame(data)
df = data.drop(columns=["Date"])
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, linecolor='black')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```

```python
data = {
    "Date": ["28-04-2023", "27-04-2023", "26-04-2023"],
    "Adj Close": [34098, 33826, 33302],}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
plt.figure(figsize=(10, 6))
plt.fill_between(df['Date'], df['Adj Close'], color="skyblue", alpha=0.4)
plt.plot(df['Date'], df['Adj Close'], color="Slateblue", alpha=0.6, linewidth=2)
plt.xlabel('Date')
plt.ylabel('Adj Close Price')
plt.title('Area Chart of Adj Close Prices Over Time')
plt.show()
```

```python
data = ({"Date": ["28-04-2023", "27-04-2023", "26-04-2023"],
        "High": [34105, 33860, 33646],
        "Low": [33108, 30210, 33126],
        "Volume": [35431, 34324, 32117]})
plt.figure(figsize=(10, 4))
plt.bar(data['Date'],data['High'],color='r', label='High')
plt.bar(data['Date'],data['Low'],color='y', label='Low')
plt.bar(data['Date'],data['Volume'],color='b',label='Volume')
plt.title('Histogram of Volume Distribution')
plt.xlabel('Volume')
plt.ylabel('Frequency')

plt.legend()
plt.show()
```

```python
data = {
    "Date": ["01-10-2023", "02-10-2023", "03-10-2023", "04-10-2023", "05-10-2023", 
             "06-10-2023", "07-10-2023", "08-10-2023", "09-10-2023", "10-10-2023"],
    "Close": [1258,29599,4007,18572,26072,29191,33600,36800,34105, 33860]}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['7-day MA'] = df['Close'].rolling(window=7).mean()
df['30-day MA'] = df['Close'].rolling(window=30).mean()
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue', marker='o', linestyle='-', alpha=0.7)
plt.plot(df['Date'], df['7-day MA'], label='7-day Moving Average', color='green', linestyle='-', linewidth=2)
plt.plot(df['Date'], df['30-day MA'], label='30-day Moving Average', color='red', linestyle='-', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Close Price with 7-day and 30-day Moving Averages')
plt.legend()
plt.show()
```
