# Bar Chart

```python
import matplotlib.pyplot as plt
titles = ["The Night Agent: Season 1", "Ginny & Georgia: Season 2","You:Season 4"]
hours_viewed = [429600000, 503000000,440600000]
plt.figure(figsize=(9,4))
plt.bar(titles, hours_viewed,color=['g', 'y','r'])
plt.xlabel('Title')
plt.ylabel('Hours Viewed')
plt.title('Hours Viewed by Each Title')
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
titles = ["The Night Agent: Season 1", "Ginny & Georgia: Season 2","You:Season 4"]
hours_viewed = [429600000, 503000000,440600000]
plt.figure(figsize=(4,4))
plt.pie(hours_viewed,labels=titles,autopct='%1.1f%%')
plt.title('Hours Viewed by Each Title')
plt.show()
```

## Line Plot

```python
import matplotlib.pyplot as plt
df=pd.DataFrame({"titles":["Ginny & Georgia: Season2","La Reina del Sur: Season3","Queen Charlotte: A Bridgerton Story","The Night Agent: Season 1"],
                "Hours Viewed":[665100000,429600000,503000000,812100000],
                "Release Date":["05-01-2023","30-12-2022","04-05-2023","23-03-2023"]})
plt.plot(df["Release Date"],df["Hours Viewed"],label='Release Date',color="red",marker="o",mfc='y',mec='b',linestyle="--")
plt.xlabel("Release Date")
plt.legend()
plt.grid()
plt.ylabel("Hours viewed")
plt.title("Trends on hours view by release date")
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
df=pd.DataFrame({"titles":["Wednesday: Season1","You: Season 4","The Glory: Season1"],
                "Hours Viewed":[507700000,440600000,622800000],
                "Release Date":["23-11-2022","09-02-2023","30-12-2022"]})
plt.scatter(df["Release Date"],df["Hours Viewed"],label='Release Date',color="red",marker="o")
plt.scatter(df["Release Date"],df["Hours Viewed"],label='mohan',color="y",marker="o",edgecolor='r')
plt.xlabel("Release Date")
plt.legend()
plt.ylabel("Hours viewed")
plt.title("Hours Viewed vs. Release Date")
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
df=({
    "Title": ["The Night Agent: Season 1", "Ginny & Georgia: Season 2", "Another Show: Season 1", "Sample Show: Season 3"],
    "Hours Viewed": [812100000, 665100000, 500000000, 300000000]})
plt.figure(figsize=(6, 6))
plt.hist(df['Hours Viewed'], bins=5, color='g', edgecolor='red')
plt.xlabel('Hours Viewed')
plt.ylabel('Frequency')
plt.title('Distribution of Hours Viewed Across Titles')
plt.show()
```

#  Stacked Bar Chart

```python
import matplotlib.pyplot as plt
import pandas as pd
df = ({"Title": ["The Night Agent: Season 1", "Ginny & Georgia: Season 2", "Another Show: Season 1"],
         "English": [812100000, 665100000, 400000000],
         "Spanish": [100000000, 80000000, 50000000],
         "French": [50000000, 30000000, 20000000]})
plt.figure(figsize=(10, 4))
plt.bar(df['Title'], df['English'], color='lightblue', label='English')
plt.bar(df['Title'], df['Spanish'], color='r', label='Spanish')
plt.bar(df['Title'], df['French'], color='y', label='French')
plt.xlabel('Title')
plt.ylabel('Hours Viewed')
plt.title('Hours Viewed Segmented by Language')
plt.legend()
plt.show()

```

```python
data = {"titles":["Wednesday: Season1","You: Season 4","The Glory: Season1"],
        "Hours Viewed":[507700000,440600000,622800000]}
df = pd.DataFrame(data)
df.boxplot(column="Hours Viewed", by='titles', grid=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Hours viewed by title')
plt.suptitle('') 
plt.xlabel('Gender')
plt.ylabel('Study Hours Per Week')
plt.show()
```
