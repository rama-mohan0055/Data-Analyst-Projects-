#Histogram

```python
data = {
    "Scheme": [
        "Kotak Corporate Bond Fund(M-IDCW)", "Kotak Corporate Bond Fund(G)",
        "Aditya Birla SL Liquid Fund(DD-IDCW)", "Aditya Birla SL Liquid Fund(G)",
        "Aditya Birla SL Money Manager Fund(W-IDCW)"],
    "Net_Asset_Value(Rs.)": [1035.21, 3133.58, 100.20, 356.66, 100.18]}
df = pd.DataFrame(data)
df['Net_Asset_Value(Rs.)'].plot(kind='hist', bins=5, color='skyblue', edgecolor='black', figsize=(10, 6))
plt.title('Distribution of Net Asset Value (Rs.)')
plt.xlabel('Net Asset Value (Rs.)')
plt.ylabel('Frequency')
plt.show()
```

```python
data = {
    "Net_Asset_Value(Rs.)": [1035.21, 3133.58, 100.20, 356.66, 100.18],
    "CAGR% 1 Year": [3.69, 3.69, 4.1, 4.17, 5.19],
    "Exp. Ratio(%)": [0.67, 0.67, 0.34, 0.34, 0.33]}
df = pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
data = {
    "Scheme": [
        "Kotak Corporate Bond Fund(M-IDCW)", "Kotak Corporate Bond Fund(G)",
        "Aditya Birla SL Liquid Fund(DD-IDCW)", "Aditya Birla SL Liquid Fund(G)",
        "Aditya Birla SL Money Manager Fund(W-IDCW)"],
    "Net_Asset_Value(Rs.)": [1035.21, 3133.58, 100.20, 356.66, 100.18]}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Net_Asset_Value(Rs.)', color='skyblue')
plt.title('Box Plot of Net Asset Value (Rs.)')
plt.ylabel('Net Asset Value (Rs.)')
plt.show()
```

```python
data = {
    "Date": ["2023-01-01", "2023-02-01", "2023-03-01", "2023-04-01", "2023-05-01"],
    "Net_Asset_Value(Rs.)": [1035.21, 1200.34, 1500.65, 1800.78, 2000.90]}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'])
plt.figure(figsize=(10, 6))
plt.fill_between(df['Date'], df['Net_Asset_Value(Rs.)'], color='skyblue', alpha=0.6)
plt.title('Cumulative Trend of Net Asset Value (Rs.) Over Time')
plt.xlabel('Date')
plt.ylabel('Net Asset Value (Rs.)')
plt.xticks(rotation=45)
plt.show()
```
